```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
X = df.drop(['Name','Ticket','Cabin','Embarked'],axis =1)
```

```python
y = df(['PassengerId','Survived','Pclass','Age','SibSp','Parch','Fare'])
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
    
plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show
```

```python
X['Cluster'] = KMeans(4).fit_predict(X)
X.groupby('Cluster').agg(['mean']).reset_index()
```
